```python
# Import necessary libraries
import lightgbm as lgb
import xgboost as xgb
from catboost import CatBoostClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
# user-defined parameters

algo_name = 'GBM,LightGBM,CatBoost,XGBoost'
is_ensemble = 'yes'
n_trees = 40
depth = 15
lr = 0.001
```

##### Train

```python
clf_gbm = GradientBoostingClassifier(n_estimators=n_trees, max_depth=depth, learning_rate=lr).fit(X,y)
clf_lgbm = lgb.LGBMClassifier(num_leaves=40,n_estimators = n_trees, max_depth = depth, learning_rate = lr).fit(X,y)
clf_cat = CatBoostClassifier(n_estimators=n_trees, depth=depth, learning_rate=lr, verbose=False).fit(X,y)
clf_xgb = xgb.XGBClassifier(n_estimators=n_trees, max_depth=depth, learning_rate=lr).fit(X,y)
```

```python
# Create the ensemble model
eclf = VotingClassifier(estimators=[('lgbm', clf_lgbm), ('xgb', clf_xgb), ('cat', clf_cat),('gbm',clf_gbm)], voting='soft')
# Fit the ensemble model
eclf.fit(X, y)
```

```python
train_data['y_pred_01'] = eclf.predict(X)
test_data['y_pred_01'] = eclf.predict(X_test)
```

```python
from sklearn.metrics import classification_report

# Train accuracy
print(classification_report(y, eclf.predict(X), labels=[0,1]))
```

```python
train_data['win_pred_score'] = eclf.predict_proba(X)[:,1]
test_data['win_pred_score'] = eclf.predict_proba(X_test)[:,1]
```

##### Getting win_pred_score corresponding to winner team

```python
train_data['win_pred_score'] = np.where((train_data['y_pred_01']==0), (1-train_data['win_pred_score']), train_data['win_pred_score'])
test_data['win_pred_score'] = np.where((test_data['y_pred_01']==0), (1-test_data['win_pred_score']), test_data['win_pred_score'])
```

#### Feature importance

```python
import numpy as np

# Assuming eclf is your VotingClassifier and X is your data
feature_importances = np.mean([
    clf.feature_importances_ for clf in eclf.estimators_
], axis=0)

df_feat_importance = pd.DataFrame({
    'feat_name': X.columns.tolist(),
    'model_feat_imp_train': feature_importances
}).sort_values(by='model_feat_imp_train', ascending=False).reset_index(drop=True).head(20)

df_feat_importance
```
